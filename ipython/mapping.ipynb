```python
df = pd.read_csv('airlines.csv')
```

```python
mapping = {'周一': "Monday",
           '周二': "Tuesday",
           '周三': "Wednesday",
           '周四': "Thursday",
           '周五': "Friday",
           '周六': "Saturday",
           '周日': "Sunday",
          }
regions = {
    '北美': "North America",
    '东亚': "East Asia",
    "东南亚": "South East Asia",
    "欧洲": "Europe",
    "中东": "Middle East",
    "非洲": "Africa"
}
flights = {}
for idx, row in df.iterrows():
    flight_num = row['航班号'].strip().split('/')[0].strip()
#     print(flight_num)
    week_of_day = row['周几'].strip().split('/')[0].strip()
#     print(week_of_day)
    origin = row['机场'].replace(')', '').split('(')[1]
    origin_city = row['机场'].split('(')[0]
#     print(origin)
    destination = row['国内机场'].replace(')', '').split('(')[1]
    destination_city = row['国内机场'].split('(')[0]
#     print(destination)
    region = regions.get(row['区域'])
    country = row['国家']
    if '和' in week_of_day:
        days = week_of_day.split('和')
        mapping_days = []
        for day in days:
            mapping_days.append(mapping.get(day))
    else:
        mapping_days = [mapping.get(week_of_day)]
    flights[flight_num] = {'origin': origin, 'destination': destination, 'date': mapping_days, "region": region, "origin_city": origin_city, "destination_city": destination_city, "country": country}

print(flights)
```
